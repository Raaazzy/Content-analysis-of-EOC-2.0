```python
def has_russian_words(s):
    russian_words = re.findall(r'[а-яА-ЯёЁ]+', str(s))
    return bool(russian_words)

def remove_non_russian(text):
    # Используем регулярное выражение, чтобы оставить только русские слова
    russian_words = re.findall(r'\b[а-яА-ЯёЁ]+\b', text)
    if not russian_words:
        return ''
    return ' '.join(russian_words)

# Создаем копию DataFrame
df5_without_english = df5.copy()

# Фильтрация строк
df5_without_english = df5_without_english[df5_without_english['Текст результата'].apply(has_russian_words)]

# Применение функции к столбцу 'Текст результата'
df5_without_english['Текст результата'] = df5_without_english['Текст результата'].apply(remove_non_russian)

# Вывод результата
df5_without_english
```

```python
import pymorphy2

def extract_verbs(sentence):
    morph = pymorphy2.MorphAnalyzer()

    words = sentence.split()
    result = []

    for word in words:
        parsed_word = morph.parse(word)[0]

        # Извлекаем только глаголы и существительные, образованные от глаголов
        if parsed_word.tag.POS in {'VERB', 'INFN'}:
            # Преобразуем в начальную форму
            lemma = parsed_word.normal_form
            result.append(lemma)

    return result

# Применяем функцию к столбцу "Текст результата"
df5_without_english['Глаголы'] = df5_without_english['Текст результата'].apply(extract_verbs)
# verbs = extract_verbs_nouns(df5['Глаголы'])

# Выводим результат
display(df5_without_english)

# Замените 'output_data.xlsx' на имя файла, в который вы хотите сохранить DataFrame
output_file = '/content/drive/MyDrive/course_work/infinitive_verb.xlsx'

# Запись DataFrame в Excel-файл
df5_without_english.to_excel(output_file, sheet_name='Лист1', index=False)
```

```python
from docx import Document

def search_in_docx(file_path, search_term):
    doc = Document(file_path)

    search_term = search_term.lower()

    for index, paragraph in enumerate(doc.paragraphs):
        if re.search(r'\b{}\b'.format(re.escape(search_term)), paragraph.text):
            max_tabulations_count = len(doc.paragraphs[index].text) - len(doc.paragraphs[index].text.lstrip('\t'))
            while index > 0:
                if doc.paragraphs[index].text.endswith('ть'):
                  word = doc.paragraphs[index].text.lstrip('@').strip()
                  if len(doc.paragraphs[index].text) - len(doc.paragraphs[index].text.lstrip('\t')) <= max_tabulations_count:
                    break
                if len(doc.paragraphs[index].text) - len(doc.paragraphs[index].text.lstrip('\t')) == 0:
                  return 'None'
                index -= 1
            result = doc.paragraphs[index].text.lstrip('@').strip()
            return result
```

```python
import pymorphy2
import pandas as pd

def extract_nouns(sentence, file_path):
    morph = pymorphy2.MorphAnalyzer()

    words = sentence.split()
    result = []

    for word in words:
        parsed_word = morph.parse(word)[0]

        # Извлекаем только существительные, образованные от глаголов
        if parsed_word.tag.POS in {'NOUN'}:
            if word.endswith('ие') or word.endswith('ье'):
                lemma = search_in_docx(file_path, word)
                # Преобразуем в начальную форму
                result.append(lemma)
    return result

# Замените 'output_data.xlsx' на имя файла, в который вы хотите сохранить DataFrame
output_file = '/content/drive/MyDrive/course_work/infinitive_noun_2.xlsx'

# Создаем пустой DataFrame
output_df = pd.DataFrame(columns=['Ключ результата', 'Текст результата', 'Отглагольные существительные'])

start_index = 9370

# Итерируем по каждой строке в исходном DataFrame
for index, row in df5_without_english.iloc[start_index:].iterrows():
    # Применяем функцию к текущей строке
    extracted_nouns = extract_nouns(row['Текст результата'], '/content/drive/MyDrive/course_work/slobr_T.docx')

    # Добавляем результат в новую строку
    output_df.loc[index] = [row['Ключ результата'], row['Текст результата'], extracted_nouns]

    # Записываем DataFrame в Excel-файл после обработки всех строк
    output_df.to_excel(output_file, sheet_name='Лист2', index=False)

# Выводим результат
display(output_df)
```

```python
import pandas as pd
from collections import Counter
import ast

# Загрузка данных из файла
df = pd.read_excel('/content/drive/MyDrive/course_work/infinitive_verb_and_noun_departments.xlsx')

# Преобразование строк в столбце "Глаголы" в списки
df['Глаголы'] = df['Глаголы'].apply(ast.literal_eval)

# Создание списка всех глаголов
all_verbs = [verb for sublist in df['Глаголы'] for verb in sublist]

# Подсчет частоты использования каждого глагола
verb_frequency = Counter(all_verbs)

# Вывод результата
print("Частота использования глаголов:")
for verb, frequency in verb_frequency.items():
    print(f"{verb}: {frequency}")

# Создание DataFrame
verb_frequency_df = pd.DataFrame(list(verb_frequency.items()), columns=['Глагол', 'Частота использования'])

# Запись DataFrame в файл Excel
verb_frequency_df.to_excel('/content/drive/MyDrive/course_work/verb_frequency.xlsx', index=False)
```

```python
from collections import defaultdict

# Словарь для хранения частоты использования глаголов по департаментам
verb_department_frequency = defaultdict(Counter)

# Для каждой строки в обновленной таблице
for index, row in df.iterrows():
    department = row['Департамент']
    verbs = row['Глаголы']
    # Обновляем словарь с частотой использования глаголов в текущем департаменте
    for verb in verbs:
        verb_department_frequency[verb][department] += 1

# Вывод результата
print("Частота использования глаголов по департаментам:")
for verb, department_freqs in verb_department_frequency.items():
    print(f"Глагол: {verb}")
    for department, frequency in department_freqs.items():
        print(f"  Департамент: {department}, Частота использования: {frequency}")

# Сохранение результатов в файл Excel
verb_department_df = pd.DataFrame([(verb, department, frequency) for verb, department_freqs in verb_department_frequency.items() for department, frequency in department_freqs.items()], columns=['Глагол', 'Департамент', 'Частота использования'])
verb_department_df.to_excel('/content/drive/MyDrive/course_work/verb_department_frequency.xlsx', index=False)
```

```python
import pandas as pd
import pymorphy2

# Создаем объект анализатора pymorphy2
morph = pymorphy2.MorphAnalyzer()

# Чтение файла Excel
df = pd.read_excel('/content/drive/MyDrive/course_work/синонимические ряды с предложениями.xlsx', sheet_name='6 уровень')

# Функция для поиска существительного после глагола и его приведения к начальной форме
def find_noun_after_verb(text, verb):
    words = text.split()
    found_verb = False
    for word in words:
        # Лемматизируем слово и проверяем, является ли оно глаголом
        lemma = morph.parse(word)[0].normal_form
        if lemma == verb:
            found_verb = True
        elif found_verb and "NOUN" in morph.parse(word)[0].tag:
            return morph.parse(word)[0].normal_form
    return None

# Применяем функцию к каждой строке датафрейма
df["Существительное"] = df.apply(lambda row: find_noun_after_verb(row["Текст результата"], row["Глагол"]), axis=1)

# Сохраняем изменения в новый документ Excel
df.to_excel("/content/drive/MyDrive/course_work/синонимические ряды с предложениями с существительными_6.xlsx", index=False)
```

```python
import pandas as pd

# Функция для сопоставления глаголов и заполнения столбца "VerbId"
def match_verbs(df, verbs_df):
    result = []
    for index, row in df.iterrows():
        verb = row['Глагол']
        word = row['Существительное']
        file_number = int(row['File'].split('_')[-1].split('.')[0])  # Извлечение номера файла из имени
        filtered_verbs = verbs_df.loc[(verbs_df['Word'] == verb) & (verbs_df['Level'] == file_number)]
        verb_id = filtered_verbs['Id'].iloc[0] if not filtered_verbs.empty else None
        result.append(verb_id)
    return result

# Чтение файлов Excel и создание новой таблицы данных
dfs = []
for i in range(1, 7):
    filename = f"/content/drive/MyDrive/course_work/синонимические ряды с предложениями с существительными_{i}.xlsx"
    df = pd.read_excel(filename, engine='openpyxl')
    df['File'] = filename  # Добавление столбца с именем файла
    dfs.append(df)

# Объединение всех таблиц в одну
df_all = pd.concat(dfs, ignore_index=True)

# Чтение файла Verbs.xlsx
verbs_df = pd.read_excel("/content/drive/MyDrive/course_work/Verbs.xlsx", engine='openpyxl')

# Сопоставление глаголов и заполнение столбца "VerbId"
df_all['VerbId'] = match_verbs(df_all, verbs_df)

# Сохранение новой таблицы данных в файл Excel
df_all.to_excel("/content/drive/MyDrive/course_work/Результат.xlsx", index=False)
```

```python
import pandas as pd
import pymorphy2

# Создаем экземпляр объекта MorphAnalyzer
morph = pymorphy2.MorphAnalyzer()

# Чтение Excel-файла
df = pd.read_excel("/content/drive/MyDrive/course_work/Nouns_unique.xlsx")

# Применение изменений к столбцу Word
def inflect_word(word):
    parsed_word = morph.parse(word)[0]
    inflected_word = parsed_word.inflect({'accs'})
    return inflected_word.word if inflected_word is not None else word

df['Word'] = df['Word'].apply(inflect_word)

# Сохранение изменений обратно в Excel-файл
df.to_excel("/content/drive/MyDrive/course_work/Nouns_unique_accs.xlsx", index=False)
```
